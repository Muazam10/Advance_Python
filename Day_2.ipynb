### **Conditions & Loops**

#### **1. Conditions**

##### **i. Even or Odd**

```python
num = int(input("Enter a number: "))
if num % 2 == 0:
    print(f"{num} is even.")
else:
    print(f"{num} is odd.")
```

##### **ii. Positive / Negative / Zero**

```python
num = int(input("Enter a number: "))
if num == 0:
    print("The number is zero.")
elif num > 0:
    print("The number is positive.")
else:
    print("The number is negative.")
```

##### **iii. Marks System**

```python
marks = int(input("Enter your marks: "))
if marks >= 90:
    grade = 'A' 
elif marks >= 80:
    grade = 'B'
elif marks >= 70:
    grade = 'C'
elif marks >= 60:
    grade = 'D'
else:
    grade = 'F'
print(f"Your grade is {grade}.")
```

##### **iv. Age Check**

```python
age = int(input("Enter your age: "))
if age < 0:
    print("Invalid age.")
elif age == 0:
    print("You are a baby.")
elif age <= 13:
    print("You are a child.")
elif age <= 20:
    print("You are a teenager.")
elif age <= 65:
    print("You are an adult.")
else:
    print("You are a senior citizen.")
```

### ***University Admission Eligibility System***

```python
# Eligibility criteria
# 1. Age must be between 18 and 25
# 2. Minimum marks: 60%
# 3. Entry test score â‰¥ 50
# 4. Gap year students require at least 70% marks


s_age = int(input("Enter Student age: "))
s_marks = int(input("Enter Intermediate marks percentage: "))
entry_test_score = int(input("Enter Entry Test score out of 100: "))
gap_year = input("Did you take a gap year? (yes/no): ").strip().lower()

if 18 <= s_age <= 25:
    if entry_test_score >= 50:
        if gap_year == "yes":
            if s_marks >= 70:
                print("Eligible for Admission")
            else:
                print("Not Eligible: Gap year students require at least 70% marks")
        else:
            if s_marks >= 60:
                print("Eligible for Admission")
            else:
                print("Not Eligible: Minimum 60% marks required")
    else:
        print("Not Eligible: Entry test score is too low")
else:
    print("Not Eligible: Age must be between 18 and 25")
```
