### **Data Types**

```python
a = 10 # integer
b = 3.14 # float
c = "Muazam" # string
d = True # boolean
print(f"Integer: {a}, Float: {b}, String: {c}, Boolean: {d}")
```

```python
num_int = 42 # integer
int_num = str(num_int) # converting integer to string
print(f"Original: {num_int} ({type(num_int)}), Converted: {int_num} ({type(int_num)})")
float_num = 3.14159 # float
str_float = int(float_num) # converting float to integer
print(f"Original: {float_num} ({type(float_num)}), Converted: {str_float} ({type(str_float)})")
str_val = "123" # string
str_to_float = float(str_val) # converting string to float
print(f"Original: {str_val} ({type(str_val)}), Converted: {str_to_float} ({type(str_to_float)})")
```

```python
name = input("Enter your name: ")
age = int(input("Enter your age: "))
height = float(input("Enter your height in inches: "))
print(f"Name: {name}, Age: {age}, Height: {height} inches: ")
print(f"Data Types - Name: {type(name)}, Age: {type(age)}, Height: {type(height)}")
```

### **Student Info Calculator**

```python
s_name = input("Enter a Name :")
s_age = int(input("Enter a Age :"))
subject_1 = int(input("Enter Subject 1 :"))
subject_2 = int(input("Enter Subject 2 :")) 
subject_3 = int(input("Enter Subject 3 :"))
total_marks = subject_1 + subject_2 + subject_3
avg_marks = total_marks / 3
print(f"Student Name: {s_name}, Age: {s_age}, Total Marks: {total_marks}, Average Marks: {avg_marks}")
```
